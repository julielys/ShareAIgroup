```python
# Step 1: Create an empty student grade dictionary
student_grades = {}

# Step 2: Add students and their initial grades
student_grades["David"] = [85, 90, 78]
student_grades["Troy"] = [92, 88, 95]
student_grades["Debby"] = [75, 80, 98]

for name in student_grades:
    print(name, student_grades[name])
print()

# Step 3: Implement operations. Add new student and grades
student_grades["Chidi"] = [88, 92, 87]

#create a new user input to add a new student and grades
while True:
    student = input("Enter new student name or type quit: ")
    if student == "quit":
        break
    grades = int(input("Enter grade 1: ")), int(input("Enter grade 2: ")), int(input("Enter grade 3: "))
    student_grades[student] = grades

print(student_grades)
print()
# Update a student's grades
student_grades["Troy"].append(92)
#%%
# Delete a student
student_del = input("Enter the name of the student you want to delete: ")
del student_grades[student_del]
print(student_grades)
print(f"{student_del}'s record has been deleted")
print()

# Step 4: Calculate and display average grades
for student, grades in student_grades.items():
    average = sum(grades) / len(grades)
    print(f"{student}'s Average Grade: {average:.2f}")
print()

# Step 5: Find the student with the highest average grade
highest_average_student = max(student_grades, key=lambda student: sum(student_grades[student]) / len(student_grades[student]))
print(f"{highest_average_student} has the highest average grade.")

print()
# Step 6: Find the student with the lowest average grade
lowest_average_student = min(student_grades, key=lambda student: sum(student_grades[student]) / len(student_grades[student]))
print(f"{lowest_average_student} has the lowest average grade.")
```
